```python
import numpy as np

def square_wave(x, frequency):
    return np.sign(np.sin(2 * np.pi * frequency * x))

# Test the function
x = np.linspace(-1, 1, 1000)
y = square_wave(x, 5)  # 5 cycles in the range -1 to 1
```

```python
plt.plot(x, y)
```
